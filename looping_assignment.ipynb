13. Check whether a number is palindrome.

```
num = int(input("enter your number = "))
temp = num
reversed = 0
while temp > 0 :
  reversed = (reversed * 10 )+(temp % 10)
  temp //= 10
if num == reversed:
   print(reversed, "the number is palidrome")
else:
   print("number is not palidrome")
```
