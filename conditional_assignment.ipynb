19. Write a program to print the day name based on day number (1–7).



```
days = int(input("enter day_number = "))
match days:
     case 1:
       print("monday")
     case 2:
       print("tuesday")
     case 3:
       print("tuesday")
     case 4:
       print("wednesday")
     case 5:
       print("thursday")
     case 6:
       print("friday")
     case 7:
       print("saturday")
     case _:
      print("invalid_input ! Please enter a number between 1 and 7." )
```
