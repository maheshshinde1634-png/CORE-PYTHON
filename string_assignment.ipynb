4. Count the number of vowels in a string.

---




```python
txt = input("enter your text = ")
count = 0
for ch in txt.lower():
  if ch in 'aeiou':
    count += 1

print("number of vowels =",count)
```

5. Count the number of consonants in a string.


```python
txt = input("enter your text = ")
count = 0
for ch in txt.lower():
  if ch.isalpha() and ch not in "aeiou":
    count += 1


print("total consonants is = ",count)
```

mini project of text analyzer

```python
text = input("enter your text here = ")
while True :
    print("--------------------text formatter------------------------")
    print("1. for convert text in lowercase ")
    print("2. for convert in capital letter")
    print("3. for convert in title style")
    print("4. for convert text in upper case")
    print("5. for exit")
    print("-----------------------------OUTPUT---------------------------")

    choices = int(input("select type of formatting = "))
    match choices :
        case 1 :
            total_char = text.lower()
            print("the lower text =", total_char)
        case 2:
            total_char = text.capitalize()
            print("the capitalized text = ",total_char)
        case 3:
            total_char = text.title()
            print("the title case= ",total_char)
        case 4:
            total_char = text.upper()
            print("the uppercase  text = ",total_char)
        case 5:
            print("exit")
            break

        case _:
            print("exit")
```
